A dictionary is a collection of key

#CREATE

```python
student = {} #empty dictionary
student["name"] = "Alice" #add new key
student["age"] = 21
print(student)
```

#READ

```python
print(student["name"])
print(student["age"])
```

#UPDATE

```python
student["name"] = "bob"
student["age"] = 25
student["grade"] = "A"
print(student)
```

#DELETE

```python
del student["grade"] #completely removes keys and values
print(student)
```

```python
removed = student.pop("age")#pop will delete the key but value is presnt
#still cannot access the value
print(student)
print(removed)
```

#Nested Dictionary

```python
student ={
    "Rida":{"Math":89,"Science":90,"English":99},
    "Prats":{"Math":80,"Science":70,"English":89},
    "Nands":{"Math":69,"Science":90,"English":99},
    "lakss":{"Math":99,"Science":70,"English":79},
}
print(student)
```

#Dictionary manipulation

```python
import pandas as pd
data = pd.DataFrame(student) #pass the dictionary to data frame and specify an index
data
```

```python
data["Average"] = data.mean(axis=1)
data
```

```python
data["Math"] = data.mean(axis=0)
#axis = 0 denotes columnwise access to the dataframe
data
#The result is not a number
```

#Data Visualization

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
data["Average"].plot(kind="bar",color="pink",edgecolor ="black")
plt.title("Average Marks of Students")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

#How to visualise Subjectwise Performance

```python
plt.figure(figsize=(8,5))
#data["Average"].plot(kind="bar",color="pink",edgecolor ="black")
for student_name in ["Rida", "Prats", "Nands", "lakss"]: # Iterate through student names (columns)
   plt.plot(data.index, data[student_name], marker ="o", label=student_name) # Plot subject scores for each student
plt.title("Subjectwise Performance by Student")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.legend() # Add a legend to identify each student's line
plt.grid(True) # Add a grid for better readability
plt.show()
```

```python
plt.figure(figsize=(8,5))
data[["Rida","Prats","Nands","lakss"]].plot(kind='box')
plt.title("Marks distribution by students")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```
