#ZeroDivisionError


```python
try:
  print(10/0)
except ZeroDivisionError:
  print("X ZerodivisonError: You cannot divide by zero.")
```

#ValueError

```python
try:
  num=int("abc")
except ValueError:
  print("ValueError: Cannot convert string to integer")
```

#typeError

```python
try:
  result= "2" + 2
except TypeError:
  print("TypeError: you tried to add strg and int")
```

#IndexError

```python
try:
  list=[1,2,3]
  print(list[5])
except IndexError:
  print("IndexError:")
```

#Exception Handeling

```python
try:
  num1 =int(input("Enter a number:"))
  num2 =int(input("Enter another number:"))
  result= num1/num2
except ZeroDivisionError:
  print("Error: You cannot divide by zero")
except ValueError:
  print("Error: please enter numbers only")
else:
  print("Result",result)
finally:
  print("program excuted successfully")
```
